## Inconsistent Data

Inconsistent data is likely due to inconsistent formatting and can be addressed by re-formatting all values in a column or row.

```python
# Look out for inconsistent data!
# The first one I notied was the Division Name column, which has some overlapping values, such as General, General Petite, and presumably other formatting issues. 
# There are also some examples, such as the very first row of the report, where Intimates is misspelled in the Division Name.
# I am guessing there are other mistakes, so I will check for all unique values.
print(df['Division Name'].unique())
print(df['Department Name'].unique())
print(df['Class Name'].unique())
```

Did you find any inconsistent data? What did you do to clean it?
